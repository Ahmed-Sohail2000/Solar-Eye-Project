## Project Scope and Objectives üéØ

A multi-class classification problem to classify the faults of solar pv modules using UAV drones and deep learning model.

This project aims to develop an AI-driven system that utilizes drone technology to capture images or videos of solar photovoltaic (PV) modules for the purpose of predicting and classifying faults. The integration of AI with drone imagery can enhance the efficiency and effectiveness of solar panel inspections, leading to improved maintenance and operational reliability.

The solareye project falls under the Computer Vision category in Deep Learning and will implement CNN model architecture.

The experimental findings prove that the proposed approach can effectively and accurately detect numerous defaults that are present on solar cell surfaces, short-circuit, fracture, and horizontal dissolution.

## 3.0 Data Preprocessing

image.png

üì¶ Preprocessing & Dataset Pipeline

```python
import os
import json
import tensorflow as tf
from sklearn.model_selection import train_test_split

# 1. Load metadata
with open(json_path, 'r') as f:
  metadata = json.load(f)

# 2. Create class-to-index mapping
classes = sorted({v['anomaly_class'] for v in metadata.values()})
class_to_idx = {cls: i for i, cls in enumerate(classes)}
print(f"Classes found: {class_to_idx}")

# # 3. Preprocessing function (resize + normalize)
# def preprocess_image(image):

#   image = tf.image.resize(image, [224, 224])
#   image = image / 255.0

#   return image

def preprocess_image(image):
  image = tf.image.resize(image, [224, 224])

  # If image has shape (H, W, 1), convert to RGB
  if image.shape[-1] == 1:
    image = tf.image.grayscale_to_rgb(image)

  image = image / 255.0
  return image

# 4. Create list of image paths and labels
image_dir = image_path
image_paths = []
labels = []

for fname in os.listdir(image_dir):

  if fname.endswith('.jpg'):

    img_id = os.path.splitext(fname)[0]
    label = metadata.get(img_id, {}).get('anomaly_class')

    if label in class_to_idx:

      image_paths.append(os.path.join(image_dir, fname))
      labels.append(class_to_idx[label])

# 5. Use train_test_split from sklearn
X_train, X_test, y_train, y_test = train_test_split(image_paths, labels, test_size=0.1, random_state=42, shuffle = True)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42, shuffle = True)

# setup augmentation
data_augmentation = tf.keras.Sequential([
    tf.keras.layers.RandomFlip("horizontal"),
    tf.keras.layers.RandomRotation(0.1),
    tf.keras.layers.RandomZoom(0.1),
    tf.keras.layers.RandomContrast(0.1)
])

# 6. Function to create TensorFlow dataset
def build_tf_dataset(paths, labels, batch_size=32, shuffle=True, training = False):

  dataset = tf.data.Dataset.from_tensor_slices((paths, labels))

  def load_img(path, label):

    image = tf.io.read_file(path)
    image = tf.image.decode_jpeg(image, channels=3)
    image = tf.image.adjust_contrast(image, 2.0)
    image = preprocess_image(image)

    # apply data augmentation to the training data
    if training:

      image = data_augmentation(image)

    return image, label

  dataset = dataset.map(load_img, num_parallel_calls=tf.data.AUTOTUNE)

  if shuffle:

    dataset = dataset.shuffle(buffer_size=1000)

  return dataset.batch(batch_size).prefetch(tf.data.AUTOTUNE)

# 7. Build datasets
train_ds = build_tf_dataset(X_train, y_train, training = True, shuffle = True)
val_ds = build_tf_dataset(X_val, y_val, shuffle=False, training = False)
test_ds = build_tf_dataset(X_test, y_test, shuffle=False, training = False)

# 8. Print sizes
print(f"Train samples: {len(X_train)}")
print(f"Validation samples: {len(X_val)}")
print(f"Test samples: {len(X_test)}")
```

## 4.0 Deep Learning Model

### 4.1 Using ResNet50V2 pre-trained DL CNN Model

```python
# import libraries
import tensorflow as tf
from tensorflow.keras import layers, models
from tensorflow.keras.applications import ResNet50V2

# define model architecture
def create_model(num_classes):
  '''using a pretrained model
  '''
  # setup random seed
  tf.random.set_seed(42)

  # load the model
  base_model = ResNet50V2(input_shape = (224, 224, 3), include_top = False, weights = 'imagenet')

  # freeze the base model layers
  base_model.trainable = False

  # create model architecture
  model = models.Sequential([
      base_model,
      layers.GlobalAveragePooling2D(),
      layers.Dense(512, activation = 'relu'),
      layers.Dropout(0.5),
      layers.Dense(num_classes, activation = 'softmax')
  ])

  # compile model
  model.compile(optimizer = tf.keras.optimizers.Adam(learning_rate = 0.001),
                loss = 'sparse_categorical_crossentropy',
                metrics = ['accuracy'])

  return model
```

```python
# initiate the model
model = create_model(num_classes = len(class_to_idx))

initial_epochs = 10

# train the model
history = model.fit(train_ds, epochs = initial_epochs, validation_data = val_ds, verbose = 1)
```

Let's view the accuracy and loss curve

```python
# view the model loss and accuracy curve
import matplotlib.pyplot as plt

# extract accuracy and loss
train_acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

train_loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(train_acc) + 1)

# plot for accuracy
plt.figure(figsize = (10, 10))
plt.subplot(2, 1, 1)

plt.plot(epochs, train_acc, label = 'Train Accuracy')
plt.plot(epochs, val_acc, label = 'Validation Accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()

# plot loss
plt.subplot(2, 1, 2)
plt.plot(epochs, train_loss, label = 'Train loss')
plt.plot(epochs, val_loss, label = 'Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()

plt.tight_layout()
```

### 4.2 Fine Tune ResNet50V2 pre-trained Deep Learning CNN Model

Time to fine-tune the ResNet50V2 model and improve the performance.

```python
# 1. Load base model
base_model = tf.keras.applications.ResNet50V2(
    include_top=False,
    input_shape=(224, 224, 3),
    weights='imagenet'
)
base_model.trainable = False  # Freeze for now

# 2. Build the full model
model = tf.keras.Sequential([
    tf.keras.layers.Input(shape=(224, 224, 3)),
    base_model,
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(len(class_to_idx), activation='softmax')
])

model.summary()
```

```python
# setup epochs to be added with the final epoch
fine_tune_epochs = 10
total_epoch = initial_epochs + fine_tune_epochs

# Unfreeze top N layers
unfreeze_layers = 60
for layer in base_model.layers[-unfreeze_layers:]:
  layer.trainable = True

# Recompile with low learning rate
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# train the model
history_fine_tune_model = model.fit(train_ds, validation_data = val_ds, epochs = total_epoch, initial_epoch = len(history.epoch), verbose = 1)
```

### 4.4 Make predictions on the test dataset

```python
class_names = ['Cell', 'Cell-Multi', 'Cracking', 'Diode', 'Diode-Multi', 'Hot-Spot', 'Hot-Spot-Multi',
               'No-Anomaly', 'Offline-Module', 'Shadowing', 'Soiling', 'Vegetation']
```

```python
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import random
import cv2

def predict_from_test(test_ds, model, class_names, num_samples=5):
    """
    Predict and display random samples from the test dataset with thermal effect visualization.

    Parameters:
    - test_ds: tf.data.Dataset, the test dataset.
    - model: Trained TensorFlow model.
    - class_names: List of class labels.
    - num_samples: Number of random samples to display from one batch.
    """

    # Take a single batch from the test dataset
    for images, labels in test_ds.take(1):
        plt.figure(figsize=(num_samples * 3, 4))  # Set figure size dynamically

        # Get batch size and select random indices to display
        batch_size = images.shape[0]
        random_indices = random.sample(range(batch_size), min(num_samples, batch_size))

        for i, idx in enumerate(random_indices):
            img = images[idx]                # Get the image at random index
            true_label = labels[idx].numpy() # Convert label tensor to integer

            # Convert grayscale images to RGB
            if img.shape[-1] == 1:
                img = tf.image.grayscale_to_rgb(img)

            # Resize image to 224x224 as required by most models (e.g., MobileNet, AlexNet, etc.)
            resized_img = tf.image.resize(img, [224, 224])
            img_batch = tf.expand_dims(resized_img, axis=0)  # Add batch dimension

            # Make prediction
            predictions = model.predict(img_batch, verbose=0)
            predicted_label = tf.argmax(predictions[0]).numpy()

            # Get predicted and true class names
            pred_class = class_names[predicted_label]
            true_class = class_names[true_label]

            # Set the title color (green if correct, red if incorrect)
            color = "green" if predicted_label == true_label else "red"

            # Convert image tensor to NumPy and scale pixel values to 0-255
            np_img = img.numpy()
            np_img = np.clip(np_img * 255, 0, 255).astype(np.uint8)

            # Apply thermal effect:
            # - If grayscale, apply colormap on the single channel.
            # - If RGB, convert to grayscale first using OpenCV, then apply colormap.
            if np_img.shape[-1] == 1:
                thermal_image = plt.cm.jet(np_img[..., 0])  # Use jet colormap
            else:
                gray_like = cv2.cvtColor(np_img, cv2.COLOR_RGB2GRAY)
                thermal_image = plt.cm.jet(gray_like)

            # Remove alpha channel and scale back to RGB (0-255)
            thermal_image = (thermal_image[..., :3] * 255).astype(np.uint8)

            # Plot the image with predicted and true labels
            plt.subplot(1, num_samples, i + 1)
            plt.imshow(thermal_image)
            plt.axis('off')
            plt.title(f"True: {true_class}\nPred: {pred_class}", color=color, fontsize=10)

        plt.tight_layout()
        plt.show()
```
